```python
import pandas as pd
```

```python
df_player_awards = pd.read_csv('awards.csv')
df_dpoy.head()
```

```python
df_dpoy = df_player_awards.loc[(df_player_awards['award']=='dpoy') & (df_player_awards['winner']), ['seas_id', 'player_id']]
df_per_100_poss = pd.read_csv('per100poss.csv')
df_per_100_poss.head()
```

```python
df_dpoy_per_100_poss = df_dpoy.merge(df_per_100_poss, how='left', on=['seas_id', 'player_id'], copy=False)
df_dpoy_per_100_poss.describe()
```

```python
df_dpoy_per_100_poss.drop(columns=['seas_id', 'player_id', 'birth_year'], inplace=True)
df_dpoy_per_100_poss.columns
```

```python
df_extremes = df_dpoy_per_100_poss[df_dpoy_per_100_poss['age'] == df_dpoy_per_100_poss['age'].min()]
df_extremes = df_extremes.append(df_dpoy_per_100_poss[df_dpoy_per_100_poss['age'] == df_dpoy_per_100_poss['age'].max()])
df_extremes = df_extremes.append(df_dpoy_per_100_poss[df_dpoy_per_100_poss['gs'] == df_dpoy_per_100_poss['gs'].min()])
```

```python
df_extremes
```

```python
df_dpoy_box_score = df_dpoy_per_100_poss[['season', 'player', 'pos', 'age', 'experience', 'tm', 'g', 'gs',
                                          'pts_per_100_poss', 'trb_per_100_poss', 'ast_per_100_poss', 'stl_per_100_poss',
                                          'blk_per_100_poss', 'tov_per_100_poss', 'pf_per_100_poss']]
df_dpoy_box_score
```

```python
df_dpoy_box_score_sum = df_dpoy_box_score[['season', 'player']]
df_dpoy_box_score_sum['p+r+a'] = df_dpoy_box_score[['pts_per_100_poss', 'trb_per_100_poss', 'ast_per_100_poss']].sum(axis=1)
```

```python
df_dpoy_box_score_sum.plot.scatter(x = 'season', y = 'p+r+a', s = 50);
```
